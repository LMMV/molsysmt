# Mutate
*Mutate aminoacids in peptide or protein.*

With the function {func}`molsysmt.build.mutate`, specific aminoacids of a peptide or a protein can be mutated. Let's first of all build a short peptide to test how this function works.

```python
molecular_system = msm.build.build_peptide('TyrGlyGlyPheMet')
```

```python
msm.view(molecular_system, standard=True)
```

```python
molecular_system = msm.build.mutate(molecular_system, mutations={1:'ALA', 2:'LYS'}, keys='group_index')
```

```python
msm.convert(molecular_system, to_form='string:aminoacids3')
```

```python
msm.info(molecular_system, element='atom')
```

```python
msm.view(molecular_system, standard=True)
```

```python
molecular_system = msm.build.mutate(molecular_system, mutations={2:'GLY', 3:'GLY'}, keys='group_id')
```

```python
msm.convert(molecular_system, to_form='string:aminoacids3')
```

```python
molecular_system = msm.build.mutate(molecular_system, mutations={'GLY':'ALA'}, keys='group_name')
```

```python
msm.convert(molecular_system, to_form='string:aminoacids3')
```
