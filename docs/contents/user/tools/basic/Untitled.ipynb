```python
def indices_serializer(indices=None, start=0, stop=50, step=1, chunk=1):
    
    output = None
    
    if indices is None:
        output = list(range(start, stop, step))
    else:
        if stop>=len(indices):
            stop=len(indices)
        output = indices[slice(start, stop, step)]
    
    if chunk>1:
        coutput = []
        chunks = len(output)//chunk
        where = 0
        for ii in range(chunks):
            coutput.append(output[where:where+chunk])
            where += chunk
        if where<len(output-1):
            coutput.append(output[where:])
        del(output)
        output=coutput
            
    
    return output
```
